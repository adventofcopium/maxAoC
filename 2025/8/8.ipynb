```python
import numpy as np
import copy

test = True
if test:
    filename = "8_test.txt"
    num_its = 10
else:
    num_its = 1000
    filename = "8_input.txt"

with open(filename) as f:
    input = f.read().splitlines()
```

```python
boxes = np.asarray([coords.split(",") for coords in input], dtype=int)

dist_list = []
for i, box1 in enumerate(boxes):
    for j, box2 in enumerate(boxes[i+1:]):
        dist_list.append((i,j+i+1,np.linalg.norm(box2-box1)))

dist_list = sorted(dist_list, key= lambda x: x[2])
```

```python
def solution(input, dist_list, part=1):
    circuit_map = dict([(i,-1) for i in range(len(input))])
    unconnected = list(range(len(input)))
    circuits = {}
    i = 0
    if part==1:
        num = num_its
    else: num = len(dist_list)

    for box1, box2, _ in dist_list[:num]:
        box1_circ = circuit_map[box1]
        box2_circ = circuit_map[box2]
        if box1_circ != -1 and box2_circ == -1:
            unconnected.remove(box2)
            circuit_map[box2] = box1_circ
            circuits[box1_circ].append(box2)
        elif box1_circ == -1 and box2_circ != -1:
            unconnected.remove(box1)
            circuit_map[box1] = box2_circ
            circuits[box2_circ].append(box1)
        elif box1_circ == -1 and box2_circ == -1:
            unconnected.remove(box1)
            unconnected.remove(box2)
            circuits[i] = [box1, box2]
            circuit_map[box1] = i
            circuit_map[box2] = i
            i += 1
        elif box1_circ == box2_circ:
            pass
        else:
            for box in circuits[box2_circ]:
                circuit_map[box] = box1_circ
            circuits[box1_circ] += copy.copy(circuits[box2_circ])
            del circuits[box2_circ]

        if part == 2 and not unconnected and len(circuits) == 1:
            x1 = np.array(boxes[box1][0],dtype=np.int64)
            x2 = np.array(boxes[box2][0],dtype=np.int64)
            print((x1*x2))
            return None
        
    if part == 1:
        lens = [len(val) for val in circuits.values()]
        score = 1
        for j in sorted(lens, reverse=True)[:3]:
            score *= j
        print(score)
        return None
        
```

```python
# 1:
solution(input, dist_list, part=1)
```

```python
# 2:
solution(input, dist_list, part=2)
```
