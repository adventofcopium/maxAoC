```python
import numpy as np

test = True
if test:
    filename = "9_test.txt"
else:
    filename = "9_input.txt"

with open(filename) as f:
    input = f.read().splitlines()
coords = np.array([i.split(",") for i in input], dtype=np.dtype("int64"))
```

```python
# 1:

max_area = 0
for i, coord1 in enumerate(coords):
    for coord2 in coords[i+1:]:
        area = (abs(coord2[0] - coord1[0]) + 1) * (abs(coord2[1] - coord1[1]) + 1)
        if area > max_area:
            max_area = area
max_area
```

```python
# 2:

# build coordinate pairs and sort by area
coord_pairs = {}
for i, coord1 in enumerate(coords):
    for coord2 in coords[i+1:]:
        x1, x2 = min(coord1[1], coord2[1]), max(coord1[1], coord2[1])
        y1, y2 = min(coord1[0], coord2[0]), max(coord1[0], coord2[0])
        area = (x2-x1+1) * (y2-y1+1)
        coord_pairs[(x1,x2,y1,y2)] = area
sorted_pairs = sorted(coord_pairs.items(), key=lambda kv: (kv[1],kv[0]), reverse=True)


# build column and row intervals from the coordinate pairs
column_intervals = {}
row_intervals = {}
for i in range(0,len(coords)-2,2):
    row_intervals[coords[i][0]] = sorted((coords[i][1],coords[i+1][1]))
    column_intervals[coords[i+1][1]] = sorted((coords[i+1][0],coords[i+2][0]))
        

def check_intersection(x1,x2,y1,y2):
    #check for intersections with boundary:

    #left&right side
    for y, (lower, upper) in row_intervals.items():
        if y1 < y < y2:
            if lower <= x1 < upper or lower < x2 <= upper:
                return False

    #bottom&top
    for x, (lower, upper) in column_intervals.items():
        if x1 < x < x2:
            if lower <= y1 < upper or lower < y2 <= upper:
                return False
    return True

for pair, area in sorted_pairs:
    if check_intersection(*pair):
        print(area)
        break
```
