```python
import numpy as np

test = False
if test:
    filename = "9_test.txt"
else:
    filename = "9_input.txt"

with open(filename) as f:
    input = f.read().splitlines()
coords = np.array([i.split(",") for i in input], dtype=np.dtype("int64"))
```

```python
# 1:

max_area = 0
for i, coord1 in enumerate(coords):
    for coord2 in coords[i+1:]:
        area = (abs(coord2[0] - coord1[0]) + 1) * (abs(coord2[1] - coord1[1]) + 1)
        if area > max_area:
            max_area = area
max_area
```
