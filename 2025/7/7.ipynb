```python
import numpy as np
with open("7_test.txt") as f:
    input = f.read().splitlines()
grid = np.asarray([[i for i in line] for line in input])
```

```python
# 1:

num_splits = 0
rays = np.asarray([False]*len(grid[0]))
start_index = np.nonzero(grid[0] == "S")[0]
rays[start_index] = True

for line in range(2,len(grid),2):
    splitters = grid[line] == "^"
    effective_splitters = np.nonzero(splitters & rays)[0]
    num_splits += len(effective_splitters)

    rays[effective_splitters] = False #destroys incoming rays
    rays[effective_splitters - 1] = True
    rays[effective_splitters + 1] = True

num_splits
```

```python
# 2:

rays = np.zeros((len(grid[0]),), dtype=np.int64)
start_index = np.nonzero(grid[0] == "S")[0]
rays[start_index] = 1

for line in range(2,len(grid),2):
    splitters = grid[line] == "^"
    effective_splitters = np.nonzero(splitters & (rays>0))[0]

    rays[effective_splitters - 1] += rays[effective_splitters]
    rays[effective_splitters + 1] += rays[effective_splitters]
    rays[effective_splitters] = 0 #destroys incoming rays

rays.sum()
```
