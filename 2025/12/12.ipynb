```python
with open("12_test.txt") as f:
    input = f.read().splitlines()
input = [line.split(" ") for line in input]
```

```python
shapes = [[a[0] for a in input[i:i+3]] for i in range(1,30,5)]
shape_vols = [sum([sum([i == "#" for i in row]) for row in shape]) for shape in shapes]
regions = [(region[0][:-1].split("x"), [int(i) for i in region[1:]]) for region in input[30:]]
```

```python
possible = 0
impossible = 0
maybe = 0
for region in regions:
    xdim = int(region[0][0])
    ydim = int(region[0][1])
    vol_region = xdim*ydim
    min_vol_pkgs = sum([num*shape_vols[i] for i, num in enumerate(region[1])])
    max_vol_pkgs = 9 * sum(region[1])
    if vol_region < min_vol_pkgs:
        impossible += 1
    elif vol_region >= max_vol_pkgs:
        possible += 1
    else:
        maybe += 1
print(possible, impossible, maybe)
```
