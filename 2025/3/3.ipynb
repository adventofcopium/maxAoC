```python
from copy import copy

test = True
if test:
    filename = "3_test.txt"
else:
    filename = "3_input.txt"

with open(filename) as f:
    input = f.read().splitlines()
input = [[int(i) for i in line] for line in input]
```

```python
# 1:

score = 0
for bank in input:
    max_first = 0
    max_second = 0
    for battery in bank[:-1]:
        if battery > max_first:
            max_first = battery
            max_second = 0
        elif battery > max_second:
            max_second = battery
    if bank[-1] > max_second:
        max_second = bank[-1]
    score += max_first * 10 + max_second
score
```

```python
# 2:

score = 0
n = 12
for bank in input:
    max_values = [0] * n
    len_bank = len(bank)
    for i, battery in enumerate(bank):
        new_max_values = copy(max_values)
        for j, max in enumerate(max_values):
            if n - 1 - j < len_bank - i and battery > max:
                new_max_values[j] = battery
                new_max_values[j+1:] = [0]*len(new_max_values[j+1:])
                break
        max_values = new_max_values
    for i, max in enumerate(max_values):
        score += 10**(n-i-1) * max
score
```
