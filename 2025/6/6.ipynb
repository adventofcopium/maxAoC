```python
import numpy as np
test = True
if test:
    filename = "6_test.txt"
else:
    filename = "6_input.txt"

with open(filename) as f:
    input = f.read().splitlines()
```

```python
# 1:

problems = []
operations = []
    
def line_to_numbers(line):
    numbers = []
    current_num = ""
    for i, char in enumerate(line):
        if char.isdigit():
            current_num += char
            if i == len(line)-1:
                numbers.append(current_num)
        elif current_num != "":
            numbers.append(int(current_num))
            current_num = ""
    return numbers


for line in input[:-1]:
    problems.append(line_to_numbers(line))
problems = np.array(problems, dtype=np.dtype("int64"))

for char in input[-1]:
    if char != " ":
        operations.append(char)


grand_total = 0
for i in range(problems.shape[1]):
    if operations[i] == "+":
        grand_total += np.sum(problems[:,i])
    else:
        grand_total += np.prod(problems[:,i])
grand_total
```

```python
# 2:

numbers_of_nrs = []
curr = 0
for char in input[-1][1:]:
    if char == " ":
        curr += 1
    else:
        numbers_of_nrs.append(curr)
        curr = 0
numbers_of_nrs.append(curr+1)

problems = [["" for _ in range(nr)] for nr in numbers_of_nrs]

for line in input[:-1]:
    i, j = 0, 0
    tot = 0
    for j, nr in enumerate(numbers_of_nrs):
        for i in range(nr):
            char = line[tot]
            if char != " ":
                problems[j][i] += char
            tot += 1
        tot += 1
        j += 1

grand_total = 0
for i, problem in enumerate(problems):
    if operations[i] == "+":
        grand_total += np.sum(np.array(problem, dtype=np.dtype("int64")))
    else:
        grand_total += np.prod(np.array(problem, dtype=np.dtype("int64")))
grand_total
```
