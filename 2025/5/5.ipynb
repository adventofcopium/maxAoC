```python
test = True
if test:
    filename = "5_test.txt"
else:
    filename = "5_input.txt"

with open(filename) as f:
    input = f.read().split("\n\n")
id_ranges, ids = input[0].splitlines(), input[1].splitlines()
id_ranges = [(int(upper), int(lower)) for (upper,lower) in [range.split("-") for range in id_ranges]]
ids = [int(id) for id in ids]
```

```python
# 1:

score = 0
for id in ids:
    for lower, upper in id_ranges:
        if id <= upper and id >= lower:
            score += 1
            break
score
```

```python
# 2:

to_remove = []
for i in range(len(id_ranges)):
    lower1, upper1 = id_ranges[i]
    for j, (lower2, upper2) in enumerate(id_ranges[i+1:]):
        if not ((upper1 < lower2) or (lower1 > upper2)):
            new_range = (min(lower1, lower2), max(upper1, upper2))
            id_ranges[j+i+1] = new_range
            to_remove.append(i)
            break
score = 0
for i, (lower, upper) in enumerate(id_ranges):
    if i not in to_remove:
        score += upper-lower+1
score
```
