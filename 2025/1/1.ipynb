```python
with open("1_test.txt") as f:
    input = f.read().splitlines()
```

```python
# 1

zeros = 0
position = 50
for rot in input:
    if rot[0] == "R":
        position += int(rot[1:])
        position %= 100
    else:
        position -= int(rot[1:])
        position %= 100
    if position == 0:
        zeros += 1
        
zeros
```

```python
#2

zeros = 0
position = 50
for rot in input:
    num_clicks = int(rot[1:])
    zero_encounters = num_clicks // 100
    zeros += zero_encounters
    num_clicks = num_clicks - zero_encounters*100

    if rot[0] == "R":
        new_pos = position + num_clicks
        if new_pos > 99:
            zeros += 1
            new_pos -= 100
    else:
        new_pos = position - num_clicks
        if new_pos < 0:
            if position > 0:
                zeros += 1
            new_pos += 100
        elif new_pos == 0:
            zeros += 1

    position = new_pos

zeros
```
