```python
with open("4_test.txt") as f:
    input = f.read().splitlines()
grid = [[i for i in line] for line in input]
```

```python
#1

from itertools import product

score = 0
for i in range(len(grid)):
    for j in range(len(grid[0])):
        if grid[i][j] == "@":
            score += 1
            num_rolls = 0
            for m, n in product((-1,0,1),(-1,0,1)):
                    a = i+m
                    b = j+n
                    if a>=0 and b>=0 and a<len(grid) and b<len(grid[0]) and grid[a][b]=="@":
                        num_rolls += 1
                        if num_rolls > 4:
                            score -= 1
                            break

score
```

```python
# 2:

def check_removable(grid):
    removables = []

    for i in range(len(grid)):
        for j in range(len(grid[0])):
            if grid[i][j] == "@":
                num_rolls = 0
                found = True
                for m, n in product((-1,0,1),(-1,0,1)):
                        a = i+m
                        b = j+n
                        if a>=0 and b>=0 and a<len(grid) and b<len(grid[0]) and grid[a][b]=="@":
                            num_rolls += 1
                            if num_rolls > 4:
                                found = False
                                break
                if found:
                    removables.append((i,j))
    return removables
```

```python
def update_grid(grid, removables):
    for i, j in removables:
        grid[i][j] = ""
    return grid
```

```python
score = 0
while True:
    removables = check_removable(grid)
    if len(removables) == 0:
        break
    score += len(removables)
    grid = update_grid(grid,removables)
score
```
