```python
with open("2_test.txt") as f:
    input = f.read().split(",")
```

```python
# 1:

score = 0

for interval in input:
    lower, upper = interval.split("-")
    for id in range(int(lower), int(upper)+1):
        id_str = str(id)
        if len(id_str) % 2 == 0:
            if id_str[:len(id_str)//2] == id_str[len(id_str)//2:]:
                score += id

score
```

```python
# 2:

score = 0

for interval in input:
    lower, upper = interval.split("-")
    for id in range(int(lower), int(upper)+1):
        id_str = str(id)

        for i in range(1, len(id_str)//2+1):
            if len(id_str)%i == 0:
                substrings = [id_str[j:j+i] for j in range(0, len(id_str), i)]
                if len(set(substrings)) == 1:
                    score += id
                    break

score
```
